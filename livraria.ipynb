```python
print('Bem vindo a livraria do Rafael Ramos')
print ()
lista_livro = []  #lista que armazenará os livros
id_global = 0  #id global para identificar os livros


def cadastrar_livro(id):  #função para cadastrar livro
    print(f'id do livro: {id}')
    nome = input('nome do livro: ')
    autor = input('autor do livro: ')
    editora = input('editora do livro: ')
    livro = {'id': id, 'nome': nome, 'autor': autor, 'editora': editora}  #cria dicionário com dados
    lista_livro.append(livro)  #adiciona o dicionário na lista


def exibir_livro(livro):  #função para formatar a saída de um livro
    print(f"id: {livro['id']}")
    print(f"nome: {livro['nome']}")
    print(f"autor: {livro['autor']}")
    print(f"editora: {livro['editora']}")
    print()  #linha em branco para separar


def consultar_livro():  #função para consultar livros
    while True:
        print("-" * 34)
        print('--------- Menu Consulta ---------')
        print('1 - consultar todos os livros')
        print('2 - consultar por id')
        print('3 - consultar por autor')
        print('4 - sair')
        opcao = input('escolha uma opção: ')

        if opcao == '1':
            for livro in lista_livro:  #mostra todos os livros
                exibir_livro(livro)

        elif opcao == '2':
            id_consulta = int(input('digite o id do livro: '))
            for livro in lista_livro:
                if livro['id'] == id_consulta:
                    exibir_livro(livro)

        elif opcao == '3':
            autor_consulta = input('digite o nome do autor: ')
            for livro in lista_livro:
                if livro['autor'].lower() == autor_consulta.lower():
                    exibir_livro(livro)

        elif opcao == '4':
            break  #sai do menu de consulta

        else:
            print('opção inválida')  # mensagem de erro


def remover_livro():  #função para remover livro
    while True:
        id_remove = int(input('digite o id do livro que deseja remover: '))
        for livro in lista_livro:
            if livro['id'] == id_remove:
                lista_livro.remove(livro)  #remove o livro da lista
                print('livro removido com sucesso.')
                return
        print('id inválido')  #mensagem de erro


#menu
while True:
    print("-" * 34)
    print('--------- Menu Principal ---------')
    print('1 - cadastrar livro')
    print('2 - consultar livro')
    print('3 - remover livro')
    print('4 - sair')

    escolha = input('escolha uma opção: ')

    if escolha == '1':
        id_global += 1
        cadastrar_livro(id_global)

    elif escolha == '2':
        consultar_livro()

    elif escolha == '3':
        remover_livro()

    elif escolha == '4':
        print('programa encerrado.')
        break

    else:
        print('opção inválida')
```
